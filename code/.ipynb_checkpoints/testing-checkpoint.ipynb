```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def phase_correlation(img_r, img_s):
    # Fourier Transform(shift transform center to image center)
    f_r = np.fft.fft2(img_r)
    
    f_s = np.fft.fft2(img_s)
    
    # SPOMF
    angle_s = np.angle(f_s)
    angle_r = np.angle(f_r)
    
    Q = np.exp(1j * (angle_s - angle_r))
    
    # inverse Fourier Transform
    delta = np.fft.ifft2(Q)
    delta = np.abs(delta)
    
    # find the index of maximum
    x0, y0 = np.unravel_index(delta.argmax(), delta.shape)
    
#     x0 = rows - x0
#     y0 = cols - y0

    return x0, y0
```

```python
img_r = cv2.imread('lena.png', 0)

rows, cols = img_r.shape
```

### Translation Experiment

```python
tx = 55.
ty = 155

M = np.float32([[1, 0, tx],
                [0, 1, ty]])

img_s = cv2.warpAffine(img_r, M, (cols, rows))

plt.subplot(121),plt.imshow(img_r, cmap='gray')
plt.title('Input Image 1'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(img_s, cmap='gray')
plt.title('Input Image 2'), plt.xticks([]), plt.yticks([])
```

```python
x0, y0 = phase_correlation(img_r=img_r, img_s=img_s)
ret = np.array([x0, y0])
print(ret)
```
